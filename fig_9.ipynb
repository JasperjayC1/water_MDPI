```python
import numpy as np
import scipy.stats as stats
import scipy.optimize as opt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('/mnt/e/Research_life/water/提取数据.xlsx',header=1)
DATA = data.iloc[:,6:9]
DATA['yr'] = np.arange(1,len(DATA)+1)
DATA.columns = ['Year', 'es', 'ea','yr']
```

```python
x = DATA['Year']
y = DATA['es']
from matplotlib import rcParams
config = {
        "font.family": 'Times new roman',
        "font.size"  : 18,
        "font.serif" : ['SimSun'],
        "mathtext.fontset" : 'stix',
        "axes.unicode_minus": False
            
            }
rcParams.update(config)
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
fig = plt.figure(figsize=(20,10), dpi=200)
ax = fig.add_subplot(111)
ax2 = ax.twinx()
# 加入置信区间
sns.regplot(x= 'Year', y='es',  data=DATA, ax=ax, 
            line_kws={'color':'#A28C37','linestyle':'--'}, 
            scatter_kws={'s':150,'alpha':1,'color':'#6C6024'})
curve1 = ax.plot(x, y, linewidth=2, alpha=1, color='#6C6024', linestyle='-', 
            label='es')
ax.tick_params(labelsize=29,pad=10)
ax.set_ylim(7.1,7.4)
ax.set_ylabel('e$_s$(hPa)',rotation=90,fontsize = 41, labelpad=14)
plt.xlim(1969,2011)
# 加入置信区间
sns.regplot(x='Year', y='ea',  data=DATA, ax=ax2,
            line_kws={'color':'#577C8A','linestyle':'--'},
            scatter_kws={'s':150,'alpha':1,'color':'#2E5C6E'})

curve2 = ax2.plot(DATA['Year'], DATA['ea'], linewidth=2, alpha=1, color='#2E5C6E', linestyle='-', 
                label='ea')
ax2.set_ylabel('e$_a$(hPa)',rotation=90,fontsize = 40, labelpad=14)
ax2.tick_params(labelsize=29,pad=10)
ax2.set_ylim(.85,1)
ax.set_xlabel('Year',fontsize = 40,labelpad=14)
# ax.text(.27,.65,'$\mathrm{y=}$'+str(round(m,4))+'$\mathrm{*x+}$' + str(round(c,3)), fontsize=30, transform=ax.transAxes)
ax.text(.27,.70,'y = 0.0042x−1.0118', c='#6C6024', fontsize=31, transform=ax.transAxes)
ax2.text(.67,.25,'y = 0.0008x−0.7313',c='#2E5C6E',fontsize=31, transform=ax.transAxes)
plt.legend([curve1[0],curve2[0]],['es','ea'],loc='upper left',prop={'size': 31, 'family': 'Times New Roman'},scatterpoints=1,frameon=False)
## 美化
# ax.tick_params(axis='x',direction='in',bottom=False)
# ax.tick_params(axis='y',direction='out')
# for spine in ['top','right']:
#     ax.spines[spine].set_color('none')
```
