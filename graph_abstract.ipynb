```python
import numpy as np
import scipy.stats as stats
import scipy.optimize as opt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('/mnt/e/Research_life/water/提取数据.xlsx',header=1)
DATA = data.iloc[:,:5]
DATA['yr'] = np.arange(1,len(DATA)+1)
DATA.columns = ['Year', 'sh', 'vpd','et','DLR','yr']
```

```python
x = DATA['Year']
y = DATA['sh']
from matplotlib import rcParams
config = {
        "font.family": 'Times new roman',
        "font.size"  : 18,
        "font.serif" : ['SimSun'],
        "mathtext.fontset" : 'stix',
        "axes.unicode_minus": False
            
            }
rcParams.update(config)
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'

fig = plt.figure(figsize=(20,24), dpi=200)
ax = fig.add_subplot(211)
# 加入置信区间
sns.regplot(x= 'Year', y='sh',  data=DATA, ax=ax, 
            line_kws={'color':'#A28C37','linestyle':'--'}, 
            scatter_kws={'s':150,'alpha':1,'color':'#6C6024'})
curve1 = ax.plot(x, y, linewidth=2, alpha=1, color='#6C6024', linestyle='-', 
            label='sh')
ax.tick_params(labelsize=29,pad=10)
# ax.set_ylim(7.1,7.4)
ax.set_ylabel('specific humidity((kg/kg)×10$^{−3}$)',rotation=90,fontsize = 28, labelpad=14)
# ax.text(.65,.24,'y = 0.0042x−1.0118', c='#6C6024', fontsize=31, transform=ax.transAxes)
# ax.text(.65,.14,'R$^2$=0.5112',c='#6C6024',fontsize=31, transform=ax.transAxes)
ax.set_xlim(1969,2011)
ax.set_xlabel('')

ax2 = ax.twinx()
# 加入置信区间
sns.regplot(x='Year', y='vpd',  data=DATA, ax=ax2, 
            line_kws={'color':'#577C8A','linestyle':'--'},
            scatter_kws={'s':150,'alpha':1,'color':'#2E5C6E'})

curve2 = ax2.plot(DATA['Year'], DATA['vpd'], linewidth=2, alpha=1, color='#2E5C6E', linestyle='-', 
                label='vpd')
ax2.set_ylabel('VPD(hpa)',rotation=90,fontsize = 28, labelpad=14)
ax2.tick_params(labelsize=29,pad=10)
# ax2.set_ylim(.85,1)
# ax2.text(.65,.24,'y = 0.0008x−0.7313',c='#2E5C6E',fontsize=31, transform=ax2.transAxes)
# ax2.text(.65,.14,'R$^2$=0.6374',c='#2E5C6E',fontsize=31, transform=ax2.transAxes)
ax2.set_xlim(1969,2011)
ax2.set_xlabel('')

ax3 = fig.add_subplot(212)
# 加入置信区间
sns.regplot(x='Year', y='et',  data=DATA, ax=ax3, 
            line_kws={'color':'#904840','linestyle':'--'},
            scatter_kws={'s':150,'alpha':1,'color':'#904840'})

curve3 = ax3.plot(DATA['Year'], DATA['et'], linewidth=2, alpha=1, color='#904840', linestyle='-', 
                label='et')
ax3.set_ylabel('ET(mm)',rotation=90,fontsize = 28, labelpad=14)
ax3.tick_params(labelsize=29,pad=10)
# ax2.set_ylim(.85,1)
# ax3.text(.1,.90,'y = 0.0008x−0.7313',c='#904840',fontsize=31, transform=ax3.transAxes)
# ax3.text(.1,.8,'R$^2$=0.0004',c='#904840',fontsize=31, transform=ax3.transAxes)
# plt.legend([curve1[0],curve2[0]],['specific humidity(kg/kg)*E-03','ea'],loc='upper left',prop={'size': 31, 'family': 'Times New Roman'},scatterpoints=1,frameon=False)
ax3.set_xlim(1969,2011)
ax3.set_xlabel('')

ax4 = ax3.twinx()
# 加入置信区间
sns.regplot(x='Year', y='DLR',  data=DATA, ax=ax4, 
            line_kws={'color':'#20604F','linestyle':'--'},
            scatter_kws={'s':150,'alpha':1,'color':'#20604F'})

curve4 = ax4.plot(DATA['Year'], DATA['DLR'], linewidth=2, alpha=1, color='#20604F', linestyle='-', 
                label='DLR')
ax4.set_ylabel('DLR(Wm$^-2$)',rotation=90,fontsize = 28, labelpad=14)
ax4.tick_params(labelsize=29,pad=10)
# ax2.set_ylim(.85,1)
# ax4.text(.1,.90,'y = 0.0008x−0.7313',c='#20604F',fontsize=31, transform=ax4.transAxes)
# ax4.text(.1,.8,'R$^2$=0.0791',c='#20604F',fontsize=31, transform=ax4.transAxes)
ax4.set_xlim(1969,2011)
plt.subplots_adjust(wspace=0.2)  # 调整水平间距，可以根据需要调整间距大小
ax4.set_xlabel('')

plt.subplots_adjust(wspace=0.34)  # 调整水平间距，可以根据需要调整间距大小
ax.legend([curve1[0],curve2[0]],['Specific Humidity','VPD'],loc='upper left',prop={'size': 24, 'family': 'Times New Roman'},scatterpoints=1,frameon=False)
ax3.legend([curve3[0],curve4[0]],['ET','DLR'],loc='upper left',prop={'size': 24, 'family': 'Times New Roman'},scatterpoints=1,frameon=False)

## 美化
# ax.tick_params(axis='x',direction='in',bottom=False)
# ax.tick_params(axis='y',direction='out')
# for spine in ['top','right']:
#     ax.spines[spine].set_color('none')
```
